##### handle emoticons

```python
#HappyEmoticons
emoticons_happy = set([
    ':-)', ':)', ';)', ':o)', ':]', ':3', ':c)', ':>', '=]', '8)', '=)', ':}',
    ':^)', ':-D', ':D', '8-D', '8D', 'x-D', 'xD', 'X-D', 'XD', '=-D', '=D',
    '=-3', '=3', ':-))', ":'-)", ":')", ':*', ':^*', '>:P', ':-P', ':P', 'X-P',
    'x-p', 'xp', 'XP', ':-p', ':p', '=p', ':-b', ':b', '>:)', '>;)', '>:-)',
    '<3'
    ])

# Sad Emoticons
emoticons_sad = set([
    ':L', ':-/', '>:/', ':S', '>:[', ':@', ':-(', ':[', ':-||', '=L', ':<',
    ':-[', ':-<', '=\\', '=/', '>:(', ':(', '>.<', ":'-(", ":'(", ':\\', ':-c',
    ':c', ':{', '>:\\', ';('
    ])
```

```python
#Emoji patterns
emoji_pattern = re.compile("["
         u"\U0001F600-\U0001F64F"  # emoticons
         u"\U0001F300-\U0001F5FF"  # symbols & pictographs
         u"\U0001F680-\U0001F6FF"  # transport & map symbols
         u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
         u"\U00002702-\U000027B0"
         u"\U000024C2-\U0001F251"
         "]+", flags=re.UNICODE)
```

```python
#combine sad and happy emoticons
emoticons = emoticons_happy.union(emoticons_sad)
```

```python
def clean_tweets(tweet): 
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(tweet)
    
    #after tweepy preprocessing the colon symbol left remain after      
    #removing mentions
    tweet = re.sub(r':', '', tweet)
    tweet = re.sub(r'‚Ä¶', '', tweet)
    
    #replace consecutive non-ASCII characters with a space
    tweet = re.sub(r'[^\x00-\x7F]+',' ', tweet)
    
    #remove emojis from tweet
    tweet = emoji_pattern.sub(r'', tweet)
    
    #filter using NLTK library append it to a string
    filtered_tweet = [w for w in word_tokens if not w in stop_words]
    filtered_tweet = []
    
    #looping through conditions
    for w in word_tokens:    
    #check tokens against stop words , emoticons and punctuations
        if w not in stop_words and w not in emoticons and w not in string.punctuation:
            filtered_tweet.append(w)
    return ' '.join(filtered_tweet)
```

```python
data_2['clean_tweet'] = data_2['text'].apply(clean_tweets)
data_2.head()
```

```python
def remove_stopwords(row):
    stop_words = set(stopwords.words('english'))   
    word_tokens = word_tokenize(row) 
    filtered_sentence = [w for w in word_tokens if not w in stop_words] 
    return ' '.join(filtered_sentence) 
```

```python
def remove_stopwords_advanced(row):
    row = row.translate(str.maketrans('', '', string.punctuation))            # remove punctuation
    row = row.replace('#','')                                                 # remove hashtag
    row = row.replace('@','')                                                 # remove tag
    stop_words = set(stopwords.words('english'))                              # get stopword list
    stop_words = stop_words.union(set(('RT', '2020', 'Resolution2020')) )     # update stop word list
    word_tokens = word_tokenize(row) 
    filtered_sentence = [w for w in word_tokens if not w in stop_words] 
    return ' '.join(filtered_sentence) 
```

```python
def get_bigrams(row):
    tokens = nltk.word_tokenize(row)
    bigrm = nltk.bigrams(tokens)
    out = []
    for item in bigrm:
        out.append(str(item[0]) + '_' + str(item[1]) )
    return out
```

```python
def get_bigrams_2(row):
    tokens = nltk.word_tokenize(row)
    bigrm = nltk.bigrams(tokens)
    out = []
    for item in bigrm:
        out.append(str(item[0]) + str(item[1]) )
    return out
```
