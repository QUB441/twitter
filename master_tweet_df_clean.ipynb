### Load sample tweets, clean them and obtain a sentiment score

```python
#get the libraries
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
import string
import json
import os
import re
```

```python
#load in all twitter files
data_sample = json.load(open('C:/py_stuff/cloud_war/planet/all_tweets.json', encoding='utf-8'))
data = data_sample['results']
data
```

```python
#clean and sijmplify the tweets into a master df

id_ls = []
user_id_ls = []
user_name_ls = []
text_ls = []
hash_ls = []
hash_ex_ls = []
retweeted_ls = []
for line in data:
    id_ls.append(line['id'])
    user_id_ls.append(line['user']['id'])
    user_name_ls.append(line['user']['screen_name'])
    retweeted_ls.append(line['retweeted'])
    if line['truncated'] == True:
        text_ls.append(line['extended_tweet']['full_text'])
    elif line['truncated'] == False:
        text_ls.append(line['text']) 
    hashtags = []    
    for hashtag in line['entities']['hashtags']:
        hashtags.append(hashtag['text'])
        print(hashtags)
        hash_ls.append(hashtags)
    ex_hashtags = []
#     for exhash in line ['extended_tweet']['entities']['hashtags']:
#         ex_hashtags.append(exhash['text'])
#         hash_ex_ls.append(ex_hashtags)
        
    
    
tweet_df = pd.DataFrame()
tweet_df['id'] = id_ls
tweet_df['user_id'] = user_id_ls
tweet_df['screen_name']= user_name_ls
tweet_df['tweet_text'] = text_ls
tweet_df['hashtag'] = pd.Series(hash_ls)
tweet_df['retweeted'] = retweeted_ls
# tweet_df['extended_hash'] = pd.series(hash_ex_ls)
print(tweet_df[['user_id', 'hashtag']])
#tweet_df.to_csv('C:/py_stuff/cloud_war/planet/tweet_data.csv')
```
